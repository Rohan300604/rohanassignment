```python
#6 identify any outlier in a numerical column of your choice.explain your approach
numerical_column=('ef_total_count')
q1=df[numerical_column].quantile(0.25)
q3=df[numerical_column].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliers=df[(df[numerical_column]<lower_bound)|(df[numerical_column]>upper_bound)]
print(f"outliers{numerical_column}':'")
print(outliers)
```
