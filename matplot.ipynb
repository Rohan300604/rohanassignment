```python
# @title Default title text
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

violin polt


```python
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Violin Plot of Petal Length by Species')
plt.show()
```

```python
#swamplot
sns.swarmplot(x='species', y='petal_length', data=iris)
plt.title('Swarm Plot of Petal Length by Species')
plt.show()
```

```python
#striplot
sns.stripplot(x='species', y='petal_length', data=iris)
plt.title('Strip Plot of Petal Length by Species')
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=iris)
plt.title('Box Plot of Petal Length by Species')
plt.show()
```

```python
#barplot
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal_length', data=iris)
plt.title('Bar Plot of Petal Length by Species')
plt.show()
```

```python
#countplot
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris)
plt.title('Count Plot of Species')
plt.show()
```

```python
#heatmap
correlation_matrix = iris.drop(columns=['species']).corr()
#create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
#show the plot
plt.title('Correlation Heatmap')
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', data=iris)
plt.title('Scatter Plot of Petal Length vs. Petal Width')
plt.show()
```

```python
#linechart
mean_value=iris.groupby('species').mean().reset_index()
plt.figure(figsize=(8, 6))
sns.lineplot( x='species', y='sepal_length', data=mean_value,markers='0')
plt.title('Line Chart of Mean Sepal Length  by Species')
plt.show()
```
